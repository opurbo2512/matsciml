<div style='font-family:monospace; color:#8B949E; font-size:12px;
line-height:1.8; margin:8px 0;'>
The <code style='color:#00E5FF;'>Material</code> class accepts a chemical formula string. Internally it uses <strong style='color:#C9D1D9;'>pymatgen</strong> to look up the crystal structure and compute all properties lazily (only when you first access them).
</div>

```python
# ── Simplest way: just give a formula ────────────────────────
mat = Material('Fe2O3')

# That's it. The object is ready.
print(type(mat))
print(repr(mat))
```

<div style='background:#0D1117; border:1px solid #21262D;

<div style='font-family:monospace; font-size:10px; color:#8B949E;
letter-spacing:2px; margin-bottom:10px;'>5 WAYS TO CREATE A MATERIAL</div>

```python
# ── From dictionary (no internet needed) ─────────────────────
mat_mgo = Material.from_dict({
    'formula': 'MgO',
    'lattice': {
        'a': 4.211, 'b': 4.211, 'c': 4.211,
        'alpha': 90, 'beta': 90, 'gamma': 90
    },
})
print(f'Created: {mat_mgo.formula}')
```

<div style='font-family:monospace; color:#8B949E; font-size:12px;

```python
mat = Material('Fe2O3')

print('── IDENTITY ──────────────────────────────')
print(f'formula        : {mat.formula}')
print(f'name           : {mat.name}')
print(f'elements       : {mat.elements}')
print(f'n_elements     : {mat.n_elements}')
print(f'crystal_type   : {mat.crystal_type}')
```

<a id='s4'></a>
<div style='background:#0D1117; border-left:4px solid #FF6D00;
border-radius:0 8px 8px 0; padding:14px 20px; margin:20px 0 6px 0;'>
<span style='font-family:monospace; font-size:10px; color:#FF6D00;

<span style='font-family:monospace; font-size:18px; color:#E6EDF3;
font-weight:700;'>Composition Attributes</span>
</div>

<div style='font-family:monospace; color:#8B949E; font-size:12px;
line-height:1.8;'>
These attributes are derived purely from the <strong style='color:#FF6D00;'>chemical formula</strong> — no crystal structure needed. They encode electronegativity, oxidation states, ionic character, and valence electrons.
</div>

```python
print('── COMPOSITION ───────────────────────────')
print(f'oxidation_states      : {mat.oxidation_states}')
print(f'electronegativity_diff: {mat.electronegativity_diff:.3f}')
print(f'ionic_character       : {mat.ionic_character:.1f}%')
print(f'valence_electrons     : {mat.valence_electrons}')
print(f'valence_electron_count: {mat.valence_electron_count:.1f}')
print(f'd_electron_count      : {mat.d_electron_count}')
print(f'avg_atomic_mass       : {mat.avg_atomic_mass:.2f} amu')
print(f'avg_ionization_energy : {mat.avg_ionization_energy:.2f} eV')
print(f'avg_electron_affinity : {mat.avg_electron_affinity:.3f} eV')
```

<div style='background:#0D1117; border:1px solid #FF6D0044;

</span>
</div>

<div style='background:#0D1117; border-left:4px solid #FFD740;

```python
print('── STRUCTURE ─────────────────────────────')
print(f'crystal_system    : {mat.crystal_system}')
print(f'space_group       : {mat.space_group}')
print(f'space_group_number: {mat.space_group_number}')
print(f'n_symmetry_ops    : {mat.n_symmetry_ops}')
print(f'is_centrosymmetric: {mat.is_centrosymmetric}')
print(f'is_polar          : {mat.is_polar}')

print('\n── LATTICE ───────────────────────────────')
lat = mat.lattice
print(f'a={lat["a"]:.3f}Å  b={lat["b"]:.3f}Å  c={lat["c"]:.3f}Å')
print(f'α={lat["alpha"]:.1f}°  β={lat["beta"]:.1f}°  γ={lat["gamma"]:.1f}°')
print(f'cell_volume       : {mat.cell_volume:.2f} Å³')
print(f'density           : {mat.density:.3f} g/cm³')
print(f'n_atoms_unit_cell : {mat.n_atoms_unit_cell}')
```

These attributes describe the <strong style='color:#00E5FF;'>local neighborhood</strong> of atoms — bond lengths, coordination number, packing. Computed using a neighbor list with 6Å cutoff.

```python
print('── ENVIRONMENT ───────────────────────────')
print(f'coordination_number : {mat.coordination_number:.1f}')
print(f'bond_length         : {mat.bond_length:.3f} Å  (mean)')
print(f'bond_length_min     : {mat.bond_length_min:.3f} Å')
print(f'bond_length_max     : {mat.bond_length_max:.3f} Å')
print(f'bond_length_std     : {mat.bond_length_std:.3f} Å')
print(f'bond_angle_mean     : {mat.bond_angle_mean:.1f}°')
print(f'packing_fraction    : {mat.packing_fraction:.3f}')
```

```python
print('── ELECTRONIC ────────────────────────────')
print(f'electronic_type   : {mat.electronic_type}')
bg_lo, bg_hi = mat.bandgap_estimate
print(f'bandgap_estimate  : {bg_lo:.1f} – {bg_hi:.1f} eV')
print(f'bandgap_ml        : {mat.bandgap_ml:.2f} eV')

# Compare metal vs semiconductor vs insulator
test_mats = [('Cu', 'metal'), ('TiO2', 'semiconductor'), ('Al2O3', 'insulator')]
print()
print(f'{"Formula":<12} {"Type":<15} {"Band gap (ML)":>15}')
print('─' * 44)
for formula, _ in test_mats:
    m = Material(formula)
    print(f'{m.formula:<12} {m.electronic_type:<15} {m.bandgap_ml:>12.2f} eV')
```

<div style='background:#0D1117; border-left:4px solid #E040FB;

<span style='font-family:monospace; font-size:10px; color:#E040FB;
letter-spacing:3px;'>SECTION 08</span><br>
<span style='font-family:monospace; font-size:18px; color:#E6EDF3;
font-weight:700;'>Methods — compare() &amp; get_neighbors()</span>

```python
# ── get_neighbors(): local atomic environment ─────────────────
mat = Material('Fe2O3')
neighbors = mat.get_neighbors(site_idx=0)

print(f'Fe atom (site 0) has {len(neighbors)} neighbors:\n')
print(f'{"Element":<10} {"Distance":>10}  {"site_idx":>8}')
print('─' * 32)
for nbr in neighbors:
    print(f"{nbr['element']:<10} {nbr['distance']:>9.3f} Å  "
          f"{nbr['site_idx']:>8}")
```

```python
# ── compare(): side-by-side property table ────────────────────
mat1 = Material('TiO2')
mat2 = Material('SnO2')   # same structure, different metal

result = mat1.compare(mat2)

print(f'{"Property":<28} {"TiO2":>10} {"SnO2":>10} {"Δ":>10}')
print('─' * 62)

for prop, vals in result.items():
    v1, v2 = vals['mat1'], vals['mat2']
    if 'diff' in vals:
        d = vals['diff']
        if isinstance(v1, float):
            print(f'{prop:<28} {v1:>10.3f} {v2:>10.3f} {d:>+10.3f}')
    else:
        tag = '✓ same' if vals['same'] else '✗ diff'
        print(f'{prop:<28} {str(v1):>10} {str(v2):>10} {tag:>10}')
```

<div style='font-family:monospace; color:#8B949E; font-size:12px;
line-height:1.8;'>
Process many materials with a simple list comprehension and build a summary DataFrame.
</div>

```python
import pandas as pd

formulas = [
    'Fe2O3', 'TiO2',  'SiC',    'Al2O3', 'GaN',
    'ZnO',   'MgO',   'BaTiO3', 'SnO2',  'CeO2'
]

materials = [Material(f) for f in formulas]

rows = []
for m in materials:
    rows.append({
        'formula'      : m.formula,
        'crystal_sys'  : m.crystal_system,
        'density'      : round(m.density, 2),
        'CN'           : m.coordination_number,
        'en_diff'      : round(m.electronegativity_diff, 2),
        'bandgap_ml'   : round(m.bandgap_ml, 2),
        'type'         : m.electronic_type,
    })

df = pd.DataFrame(rows)
df.style.background_gradient(subset=['bandgap_ml'], cmap='YlOrRd')
```
